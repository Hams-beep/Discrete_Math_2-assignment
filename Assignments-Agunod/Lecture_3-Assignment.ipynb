### $\textbf{Bipartite Graph}$


```python
import matplotlib.pyplot as plt
import networkx as nx

Bi_graph = nx.Graph()

Bi_graph.add_nodes_from([0,1,2,3], bipartite=0)
Bi_graph.add_nodes_from([4,5,6,7,8], bipartite=1)
Bi_graph.add_edges_from([(0,4),(0,5),(0,6),(0,7),(0,8), 
                         (1,4),(1,5),(1,6),(1,7),(1,8),
                         (2,4),(2,5),(2,6),(2,7),(2,8),
                         (3,4),(3,5),(3,6),(3,7),(3,8)
                         ])
top = nx.bipartite.sets(Bi_graph)[0]
pos = nx.bipartite_layout(Bi_graph, top)
nx.draw(Bi_graph, pos=pos, with_labels=True)
plt.show()
```

This is a Bipartite Graph because all the vertices in one set are connected to all the vertices in the other set and the vertices in the same set are not connected to each other.

### $\textbf{Travelling a Graph}$


##### $\textbf{Walk}$

```python
import matplotlib.pyplot as plt
import networkx as nx
seed = 45
# Use seed for reproducibility
G = nx.Graph()

# Add nodes
G.add_nodes_from([1,2,3,4,5,6])
# Add self-loops to the remaining nodes
edgelist = [(1,2),(1,5),(5,2),(1,3),(4,2),(4,6),(5,6),(3,4),(6,2),(2,3)]
walk = [(1,5),(5,6),(5,2),(2,4),(4,3),(3,2)]
G.add_edges_from(edgelist)
#pos = nx.spring_layout(G, seed=seed)  # Seed for reproducible layout
pos = nx.spring_layout(G, seed=seed)
nx.draw(G, pos=pos,with_labels=True)
nx.draw_networkx_edges(G, pos=pos, edgelist=walk, edge_color='red', width=2)
plt.show()
```

A walk is an alternating sequence of vertices and edges

The walk above from $v_1$ to $v_2$ has the route:<br>
$v_1 \to v_5 \to v_6 \to v_5 \to v_2 \to v_4 \to v_3 \to v_2$

### $\textbf{Euler's Circuit}$


A Euler's circuit in a graph contains every vertex at least once and every edge exactly once and must start and end with the same vertex.

```python
import matplotlib.pyplot as plt
import networkx as nx
seed = 20
# Use seed for reproducibility
G = nx.Graph()

# Add nodes
G.add_nodes_from([1,2,3,4,5])
# Add self-loops to the remaining nodes
edgelist = [(1,2),(3,2),(1,3),(4,1),(3,4),(5,3),(5,1)]

G.add_edges_from(edgelist)

#pos = nx.spring_layout(G, seed=seed)  # Seed for reproducible layout
pos = nx.spring_layout(G, seed=seed)
nx.draw(G, pos=pos,with_labels=True)
plt.show()
```

The Euler's circuit in this graph would be:

$v_2 \to v_3 \to v_4 \to v_1 \to v_5 \to v_3 \to v_1 \to v_2$

### $\textbf{Hamiltonian Circuit}$


A Hamiltonian Circuit in a graph is a circuit where every vertex appears exactly once except for the first and last vertex.

```python
import matplotlib.pyplot as plt
import networkx as nx
seed = 45
# Use seed for reproducibility
G = nx.Graph()

# Add nodes
G.add_nodes_from([1,2,3,4,5,6,7])
# Add self-loops to the remaining nodes
edgelist = [(1,2),(1,5),(5,2),(1,3),(4,2),(4,6),(5,6),(3,4),(6,2),(2,3),(7,4),(7,3)]
circuit = [(5,6),(6,4),(4,7),(7,3),(2,3),(2,1),(1,5)]
G.add_edges_from(edgelist)
#pos = nx.spring_layout(G, seed=seed)  # Seed for reproducible layout
pos = nx.spring_layout(G, seed=seed)
nx.draw(G, pos=pos,with_labels=True)
nx.draw_networkx_edges(G, pos=pos, edgelist=circuit, edge_color='red', width=2)
plt.show()
```

The Hamiltonian Circuit above has the route:

$v_4 \to v_7 \to v_3 \to v_2 \to v_1 \to v_5 \to v_6 \to v_4$


### $\textbf{Travelling Salesman Problem}$


The traveling salesman problem is about finding the shortest distance circuit in a weighted graph.

```python
import matplotlib.pyplot as plt
import networkx as nx
seed = 55
# Use seed for reproducibility
G = nx.Graph()

# Add nodes
G.add_nodes_from(["A","B","C","D","E"])
# Add self-loops to the remaining nodes
edgelist = [("A","B"),("A","C"),("A","D"),("B","C"),("B","D"),("C","D"),("A","C"),("B","D"),("E","A"),("E","D"),("E","B")]
G.add_edges_from(edgelist)
pos = nx.planar_layout(G)  # Seed for reproducible layout
nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels={('A', 'B'): '20', 
                 ('A', 'C'): '50', 
                 ('A', 'D'): '40',
                 ('B', 'C'): '30',
                 ('B', 'D'): '35',
                 ('C', 'D'): '25',
                 ("E","A"): '20',
                 ("E","D"): '10',
                 ("E","B"): '60'},
    font_color='red'
)
nx.draw(G, pos=pos,with_labels=True)
nx.draw(
    G, pos, edge_color='black', width=1, linewidths=1,
    node_size=500, node_color='pink', alpha=0.9,
    with_labels=True)

plt.show()
```

The possible Hamiltonian circuits in this graph are:

$A \to B \to C \to D \to E \to A = 20 + 30 + 25 + 10 + 20 = 105$<br>
$A \to B \to E \to D \to C \to A = 20 + 60 + 10 + 25 + 50 = 165$<br>
$A \to D \to E \to B \to C \to A = 40 + 10 + 60 + 30 + 50 = 190$<br>
$A \to D \to C \to B \to E \to A = 40 + 25 + 30 + 60 + 20 = 175$<br>
$A \to C \to D \to B \to E \to A = 50 + 25 + 35 + 60 + 20 = 190$<br>
$A \to C \to B \to D \to E \to A = 50 + 30 + 35 + 10 + 20 = 145$

The Best route in this graph is $A \to B \to C \to D \to E \to A$.
